```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
import kagglehub
import os
import shutil
```

```python
# Download the dataset using kagglehub
path = kagglehub.dataset_download("uciml/human-activity-recognition-with-smartphones")

# Specify your desired save directory
save_directory = "/content/HumanActivity"

# Ensure the directory exists
os.makedirs(save_directory, exist_ok=True)

# Copy extracted files to the desired directory
for file_name in os.listdir(save_directory):
    shutil.move(os.path.join(save_directory, file_name), save_directory)

# Load the dataset into a pandas DataFrame
data_set = pd.read_csv(f"{save_directory}/train.csv")

# Display the first few rows of the dataset
print(data_set.head())
```

```python
plt.figure(figsize = (10, 5))
data_set.Activity.value_counts().plot.pie(explode = (0.1, 0, 0, 0, 0, 0), autopct = "%2.2f%%")
```

```python
Acc = 0
Gyro = 0
other = 0
for signals in data_set.columns:
  if 'Acc' in signals:
    Acc = Acc + 1
  elif 'Gyro' in signals:
    Gyro = Gyro + 1
  else:
    other = other + 1

plt.figure(figsize = (7, 5))
sns.barplot(x = ['Accelerometer ','Gyroscope ','other'], y = [Acc,Gyro,other])
```

```python
new_cols = data_set.columns
new_cols = new_cols.str.replace('[()]', '')
new_cols = new_cols.str.replace('[-]', '')
new_cols = new_cols.str.replace('[,]', '')
data_set.columns = new_cols
print(data_set.columns)
```

```python
plt.figure(figsize = (7, 5))
sns.boxplot(x = 'Activity', y ='tBodyAccMagmean', data = data_set, showfliers = False);
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.boxplot(x = 'Activity', y ='angleXgravityMean', data = data_set, showfliers = False);
plt.xticks(rotation = 90)
plt.show()
```

```python
from sklearn.manifold import TSNE
X_for_tsne = data_set.drop(['subject','Activity'], axis = 1)
tsne = TSNE().fit_transform(X_for_tsne)
plt.figure(figsize = (12,8))
sns.scatterplot(x = tsne[:,0], y = tsne[:,1], hue = data_set["Activity"], palette = "bright")
```

```python
X = data_set.drop('Activity', axis = 1)
Y = data_set['Activity']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

```python
model = MLPClassifier()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
print(accuracy_score(y_test, predictions))
```
